```python
def show_triplets_history(dataframe, frame_from, length_frames):
    data = dataframe[frame_from : frame_from+length_frames]
    for frame in range (len(data)) :
        x5 = data['X5'].iloc[frame]
        x6 = data['X6'].iloc[frame]
        x7 = data['X7'].iloc[frame] 
        triplet_x = []
        triplet_x.append(x5)
        triplet_x.append(x6)
        triplet_x.append(x7)
        
        y5 = data['Y5'].iloc[frame]
        y6 = data['Y6'].iloc[frame]
        y7 = data['Y7'].iloc[frame] 
        triplet_y = []
        triplet_y.append(y5)
        triplet_y.append(y6)
        triplet_y.append(y7)
        
        cur_label = 'triplet_№' + str(frame+1)
        plt.plot(triplet_x, triplet_y, label=cur_label, linewidth=1, markersize=2,  marker='o')
    
    # Naming the x-axis, y-axis and the whole graph
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("triplets location")

    # Adding legend, which helps us recognize the curve according to it's color
    plt.legend()
      
    #plt.xlim=[500,1600]
    plt.grid()
    plt.show()
    
    
show_triplets_history(hour_data, frame_from=400, length_frames=10)
```
